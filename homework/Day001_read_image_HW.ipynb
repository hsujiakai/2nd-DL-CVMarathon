思考一下我們前面有提到圖片是矩陣，但維度可能會不一樣
例如灰階圖只有兩個維度，RGB 彩圖則有 3 個維度

假如今天我們把 RGB 3 個維度拆開來看會有甚麼不同的效果呢？

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../data/lena.png', cv2.IMREAD_COLOR)
```

```python
(B,G,R) = cv2.split(img)
G[:,:] = 0
B[:,:] = 0
merged = cv2.merge([R,G,B])
plt.axis('off')
plt.title('Channel 1')
plt.imshow(merged)
```

```python
(B,G,R) = cv2.split(img)
R[:,:] = 0
B[:,:] = 0
merged = cv2.merge([R,G,B])
plt.axis('off')
plt.title('Channel 2')
plt.imshow(merged)
```

```python
(B,G,R) = cv2.split(img)
R[:,:] = 0
G[:,:] = 0
merged = cv2.merge([R,G,B])
plt.axis('off')
plt.title('Channel 3')
plt.imshow(merged)
```
